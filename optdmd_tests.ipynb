# Simple Example

```python
from pydmdz import OptDMD
```

```python
## load data
import scipy.io
xdata=scipy.io.loadmat('simple_example.mat')['xdata']
t=scipy.io.loadmat('simple_example.mat')['ts']

print(xdata.shape)
print(t.shape)
print(xdata[-1,-1])
```

```python
r = 3
optdmd = OptDMD(xdata, t, r)
```

```python
optdmd.fit(projected=False, eigs_guess=None)
print(optdmd.eigs)
optdmd.plot_analysis()
```

```python
optdmd.fit(projected=True, eigs_guess=None)
print(optdmd.eigs)
optdmd.plot_analysis()
```

```python
import numpy as np
eigs_guess = np.random.randn(3)
eigs_guess
```

```python
optdmd.fit(projected=True, eigs_guess=eigs_guess)
print(optdmd.eigs)
optdmd.plot_analysis()
```
